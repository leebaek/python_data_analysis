# 1. DataFrame 병합하기

```python
import pandas as pd
```

### 데이터프레임 생성

```python
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'],
                   'B' : ['B0', 'B1', 'B2', 'B3'],
                   'C' : ['C0', 'C1', 'C2', 'C3'],
                   'D' : ['D0', 'D1', 'D2', 'D3']},
                  index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A' : ['A4', 'A5', 'A6', 'A7'],
                   'B' : ['B4', 'B5', 'B6', 'B7'],
                   'C' : ['C4', 'C5', 'C6', 'C7'],
                   'D' : ['D4', 'D5', 'D6', 'D7']},
                  index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A' : ['A8', 'A9', 'A10', 'A11'],
                   'B' : ['B8', 'B9', 'B10', 'B11'],
                   'C' : ['C8', 'C9', 'C10', 'C11'],
                   'D' : ['D8', 'D9', 'D10', 'D11']},
                  index=[8, 9, 10, 11])
```

### 1.1 concat명령 : 데이터를 열 방향으로 단순 병합

```python
# 단순 병합
result = pd.concat([df1, df2, df3])
result
```

```python
# 다중index(level) 생성 병합
result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
result
```

```python
# level 확인
result.index.get_level_values(0)
result.index.get_level_values(1)
```

### df4생성 후 df1과 병합

```python
df4 = pd.DataFrame({'B' : ['B2', 'B3', 'B6', 'B7'],
                   'D' : ['D2', 'D3', 'D6', 'D7'],
                   'F' : ['F2', 'F3', 'F6', 'F7']},
                  index=[2, 3, 6, 7])
result = pd.concat([df1, df4], axis=1) # axis=1은 오른쪽으로 병합하겠다는 의미 (<->axis=0)
result
```

```python
# join='inner' : 공통된 index로 합치고, 공통되지 않은 index의 데이터를 버림
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
# join_axes=[df1.index] : df1의 index에 맞춰서 병합
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
# ignore_index=True : 두 df의 index를 무시하고 새로운 index를 생성
result = pd.concat([df1, df4], ignore_index=True)
result
```

### 데이터 프레임 생성

```python
left = pd.DataFrame({'key' : ['K0', 'K4', 'K2', 'K3'],
                    'A' : ['A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key' : ['K0', 'K1', 'K2', 'K3'],
                     'C' : ['C0', 'C1', 'C2', 'C3'],
                     'D' : ['D0', 'D1', 'D2', 'D3']})
```

### 1.2 merge 명령

```python
# 'key'를 기준으로 두 데이터프레임을 병합, 공통된 key에 대해서만 병합한다.
pd.merge(left, right, on='key')
```

```python
# 한쪽 데이터를 기준으로 key를 잡아서 병합
pd.merge(left, right, how='left', on='key')
pd.merge(left, right, how='right', on='key')
```

```python
# 병합한 모든 데이터의 결과
pd.merge(left, right, how='outer', on='key')
```

```python
# 병합 후 공통된 요소만 출력
pd.merge(left, right, how='inner', on='key')
```
