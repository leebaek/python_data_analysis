```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline

tips = sns.load_dataset("tips")
tips.head()
```

### boxplot

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set1")
# hue옵션으로 구분이 가능하다.
# palette : 색상 조정
plt.show()
```

### lmplot

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
# data=tips : tips라는 데이터를 쓰겠다.
plt.show()
```

###### * 직선으로 regression도 그려주고, 유효범위도 ci로 잡아준다.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set2", size=7)
plt.show()
```

###### * new data

```python
flights = sns.load_dataset("flights")
flights.head()
```

###### * pivot으로 월별, 연도별로 구분

```python
flights = flights.pivot("month", "year", "passengers")
flights
```

### heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d") # annot : 네모칸에 값을 넣느냐 마느냐
plt.show()
```
